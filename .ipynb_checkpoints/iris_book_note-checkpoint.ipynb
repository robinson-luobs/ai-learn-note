## Ch_02_安装使用Anaconda

### 查看 Python 第三方库版本号

* 在Anaconda Prompt，用 conda list，可以调取所有安装 Python 库的
版本号
* 在 Anaconda Prompt，用 pip show 库名，比如 pip show numpy，来调取某个特
定 Python 库信息
* 在 JupyterLab 中，用!pip list 查看已经安装的所有 Python 库版本号
* 在 JupyterLab 中，!dir 可以调取当前文件的目录（Windows系统），mac系统用!pwd
* 用Python查看特定库的版本号，如下代码

```python
import numpy
print('numpy:%s'% numpy.__version__)
```

### 安装、卸载、更新库

* 安装 Python 库,在 Anaconda Prompt 中键入 pip install library_name
* 在 JupyterLab 中，!pip install library_name（不建议初学者使用）
* 源代码安装库，下载或克隆压缩文件，利用 python setup.py install 命令安装（不建议初学者使用）
* 卸载 Python 库,在 Anaconda Prompt 中键入 pip uninstall library_name
* 更新 Python 库,在 Anaconda Prompt 中键入 pip install library_name --upgrade

### 0.导入库
1) NumPy：用于科学计算和数据分析的库，提供了高效的多维数组和矩阵运算功能。
2) Pandas：用于数据处理和分析的库，提供了高效的数据结构和数据操作功能。
3) Matplotlib：用于数据可视化和绘图的库，提供了各种绘图函数和图形展示方式。 
4) Seaborn是一个基于matplotlib的Python数据可视化库，提供了更高级的接口和更美观的默认样式，使得绘制统计图形更加方便和美观。  
4) Plotly：一款交互式的数据可视化工具，可用于创建丰富的图形和可视化应用程序，支持多种编程语言和平台。

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as ans
```

### 1.代码 

#### 1.1生成等差数列

```python
a0 = 1

n = 10

d = 2

a_array = np.arange(a0, a0 + n*d ,d)
print('打印等差数列')
a_array
```

```python
b_array = np.arange(0, 10)
b_array
```

```python
c_array = np.arange(0, 10, 2)
c_array
```

#### 1.2生成二元函数数据

```python
x_array = np.linspace(1, 10, 10)
x_array
```

```python
y_array = np.linspace(1, 5, 5)
y_array
```

```python
X, Y = np.meshgrid(x_array, y_array)
```

```python
X
```

```python
Y
```

```python
plt.plot(X, Y, marker='o', color='k', linestyle='none')
plt.show()
```

其他常用标记样式 

plt.plot(xv, yv, marker='o')     # 圆圈

plt.plot(xv, yv, marker='s')     # 正方形

plt.plot(xv, yv, marker='^')     # 三角形

plt.plot(xv, yv, marker='x')     # 叉号

颜色缩写

plt.plot(xv, yv, color='k')      # 黑色 (black)

plt.plot(xv, yv, color='r')      # 红色 (red)

plt.plot(xv, yv, color='g')      # 绿色 (green)

plt.plot(xv, yv, color='b')      # 蓝色 (blue)

plt.plot(xv, yv, color='c')      # 青色 (cyan)

plt.plot(xv, yv, color='m')      # 品红 (magenta)

plt.plot(xv, yv, color='y')      # 黄色 (yellow)

plt.plot(xv, yv, color='w')      # 白色 (white)

或者使用全称 

plt.plot(xv, yv, color='black')

plt.plot(xv, yv, color='red')

plt.plot(xv, yv, linestyle='none')      # 无线条（纯散点）

plt.plot(xv, yv, linestyle='-')         # 实线（默认）

plt.plot(xv, yv, linestyle='--')        # 虚线

plt.plot(xv, yv, linestyle=':')         # 点线

plt.plot(xv, yv, linestyle='-.')        # 点划线
