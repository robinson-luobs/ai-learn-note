## Ch_02_安装使用Anaconda

### 查看 Python 第三方库版本号

* 在Anaconda Prompt，用 conda list，可以调取所有安装 Python 库的
版本号
* 在 Anaconda Prompt，用 pip show 库名，比如 pip show numpy，来调取某个特
定 Python 库信息
* 在 JupyterLab 中，用!pip list 查看已经安装的所有 Python 库版本号
* 在 JupyterLab 中，!dir 可以调取当前文件的目录（Windows系统），mac系统用!pwd
* 用Python查看特定库的版本号，如下代码

```python
import numpy
print('numpy:%s'% numpy.__version__)
```

### 安装、卸载、更新库

* 安装 Python 库,在 Anaconda Prompt 中键入 pip install library_name
* 在 JupyterLab 中，!pip install library_name（不建议初学者使用）
* 源代码安装库，下载或克隆压缩文件，利用 python setup.py install 命令安装（不建议初学者使用）
* 卸载 Python 库,在 Anaconda Prompt 中键入 pip uninstall library_name
* 更新 Python 库,在 Anaconda Prompt 中键入 pip install library_name --upgrade

### 查看库的健康情况

[https://snyk.io/advisor/python/scoring](https://snyk.io/advisor/python/scoring)
